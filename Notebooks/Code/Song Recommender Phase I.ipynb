```python
def scrape_hot100():
    url = "https://www.billboard.com/charts/hot-100/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    song_elements = soup("div", class_="o-chart-results-list-row-container")
    #print("song_elements:", song_elements)
    
    songs = []
    for element in song_elements:
        song = {}
        song["title"] = element.find("h3").text.strip()
        song["artist"] = element.find("h3").find_next("span").text.strip()
        songs.append(song)
    
    #print("songs:", songs)
    
    df = pd.DataFrame(songs)
    return df

df = scrape_hot100()
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df.index = range(1, 101)
df.head()
```

```python
df.to_csv("top_100.csv", index=False)
```

```python
def get_uri(sp, title, artist_name, limit=5):
    results = sp.search(q=title, limit=limit)
    for i in range(len(results)): 
        ar=results['tracks']['items'][i]['album']['artists'][0]['name']
        if (ar in artist_name) or (artist_name in ar):
            return results['tracks']['items'][i]['uri']
        return None 
```

```python
def get_audio_features(df):
    output = pd.DataFrame()
    for i in range (len(df['uri'])-1): 
        my_dict = sp.audio_featuers(df.loc[i,['uri']])[0]
        print(my_dict)
        my_dict_i= {key : my_dict[key] for key in list(my_dict.keys())}
        my_dict_i['title'] = df['title'][i]
        my_dict_i['artist'] = df['artist'][i]
        output = output.append(my_dict_i, ignore_index=True)
    return output
```
